```python
import numpy as np

#Escriba una función que calcule la diferencia entre el mínimo y máximo valor de un array dado(sin importa la dimensión del array)
def dif_min_max(array):

    dif = np.max(array) - np.min(array)

    return dif

#Escriba una función que calcule el determinante de un array cuadrado(el cual será de dimensión NxN)
def det_array(array):

    if array.dtype == 'int64':
      determinante = int(np.linalg.det(array))
    else:
      determinante = np.linalg.det(array)

    return determinante

#Escriba una función que normalice una matriz de NxN
#Normalizar: todos los valores de la matriz deben estar entre cero y uno
def normalizar(array):

    dimensiones = array.shape

    if np.array_equal(array,np.zeros(dimensiones)):
      matriz_normalizada = array
    else:
      media = np.mean(array)
      desv_estandar = np.std(array)
      matriz_normalizada = (array - media)/desv_estandar

    return matriz_normalizada


#Escriba una función que encuentre el valor más frecuente de un array(de cualquier dimension)
def mas_frec(array):

    valores, conteo = np.unique(array, return_counts=True)
    conteo_max = max(conteo)
    indice_max = np.argmax(conteo)

    return valores[indice_max]


#Una matriz A es antisimétrica si para todo par de índices i y j se cumple que a[i, j] == -a[j, i] (note el signo menos).
#Escriba la función es_antisimetrica(a) que indique si el array cuadrado a es antisimétrica o no.
def es_antisimetrica(array):

    return  np.allclose(array, -np.transpose(array))


#Una matriz A es simétrica si para todo par de índices i y j se cumple que a[i, j] == a[j, i].
#Escriba la función es_simetrica(a) que indique si el array cuadrado a es simétrica o no.
def es_simetrica(array):

    return np.allclose(array, np.transpose(array))

#Un cuadrado mágico es una disposición de números naturales en una tabla cuadrada, de modo que las sumas de cada columna, de cada fila y de cada diagonal son iguales.
#Escriba una función que entregue si un arreglo cuadrado de NxN es o no mágico(True o False)
def es_cuadrado_magico(array):

    Suma = np.sum(array[0, :])

    return (
        all(np.sum(array, axis=0) == Suma) and
        all(np.sum(array, axis=1) == Suma) and
        all(np.sum(np.diag(array)) == Suma) and
        all(np.sum(np.diag(np.fliplr(array))) == Suma)
          )



#Usando la clase CustomNode, complete el metodo _add(self, data, node) y _search(self, data, node) de la clase BinaryTree:
class CustomNode:
    def __init__(self, data=None, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

class BinaryTree:
    def __init__(self):
        self.root = None

    def add(self, data):
        """Método público para agregar un nodo al árbol."""
        if self.root is None:
            self.root = CustomNode(data)
        else:
            self._add(data, self.root)

    def _add(self, data, node):
        """Método privado para manejar la lógica recursiva de agregar un nodo."""
        #Agregar código aqui!
        if data < node.data:
             return self._add(data, node.left)
        elif data > node.data:
            return self._add(data, node.right)
        elif node == None:
            node = CustomNode(data)

    def search(self, data):
        """Método público para buscar un nodo en el árbol."""
        return self._search(data, self.root)

    def _search(self, data, node):
        """Método privado para manejar la lógica recursiva de buscar un nodo."""
        #Agregar código aqui!
        if node is None:
            return node
        elif data < node.data:
            return self._search(data, node.left)
        elif data > node.data:
            return self._search(data, node.right)
        else
            return True


    def remove(self, data):
        """Método público para eliminar un nodo del árbol."""
        self.root = self._remove(data, self.root)

    def _remove(self, data, node):
        """Método privado para manejar la lógica recursiva de eliminar un nodo."""
        if node is None:
            return node
        if data < node.data:
            node.left = self._remove(data, node.left)
        elif data > node.data:
            node.right = self._remove(data, node.right)
        else:
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            min_larger_node = self._get_min(node.right)
            node.data = min_larger_node.data
            node.right = self._remove(min_larger_node.data, node.right)
        return node

    def _get_min(self, node):
        """Método privado para encontrar el nodo con el valor mínimo en un subárbol."""
        current = node
        while current.left is not None:
            current = current.left
        return current
```
