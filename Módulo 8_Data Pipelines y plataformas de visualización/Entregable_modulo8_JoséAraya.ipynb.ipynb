```python
import pandas as pd
from google.colab import auth
from google.cloud import bigquery
auth.authenticate_user()
from pandas_gbq import to_gbq
```

```python
confirmed_global = pd.read_csv('/content/Covid19_confirmed_global.csv')
deaths_global = pd.read_csv('/content/Covid19_deaths_global.csv')
recovered_global = pd.read_csv('/content/Covid19_recovered_global.csv')
```

```python
confirmed_global.head()
```

```python
df1 = pd.melt(confirmed_global,id_vars= ('Country/Region','Lat','Long','Province/State'),var_name='Date',value_name='Confirmed')
df1.head()
```

```python
nulos_por_columna = df1.isnull().sum()
print(nulos_por_columna)
```

```python
deaths_global.head()
```

```python
df2 = pd.melt(deaths_global,id_vars= ('Country/Region','Lat','Long','Province/State'),var_name='Date',value_name='Deaths')
df2.head()
```

```python
nulos_por_columna = df2.isnull().sum()
print(nulos_por_columna)
```

```python
recovered_global.head()
```

```python
df3 = pd.melt(recovered_global,id_vars= ('Country/Region','Lat','Long','Province/State'),var_name='Date',value_name='Recovered')
df3.head()
```

```python
nulos_por_columna = df3.isnull().sum()
print(nulos_por_columna)
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df2.shape[0] - df3.shape[0])
```

```python
df = df1.merge(df2, on=('Country/Region','Date','Lat','Long','Province/State'), how='left')
df.head()
```

```python
df = df.merge(df3, on=('Country/Region','Date','Lat','Long','Province/State'), how='left')
df.head()
```

```python
df = df[['Country/Region','Province/State','Lat','Long','Date','Confirmed','Recovered','Deaths']]
```

```python
nulos_por_columna = df.isnull().sum()
print(nulos_por_columna)
```

```python
#Asignamos valor "no state" a nulos de Province/State
df['Province/State'] = df['Province/State'].fillna('No State')

#Asignamos valor 0 a recovered nulos y cambiamos variable a entero
df['Recovered'] = df['Recovered'].fillna(0)
df['Recovered'] = df['Recovered'].astype(int)
```

```python
df.rename(columns={'Country/Region':'Country', 'Province/State':'State'}, inplace=True)
df.head()
```

```python
#Configurar los parámetros para importar datos
project_id = 'fiery-bedrock-442322-a0'
dataset_id = 'diplom_usach_2024_JAE'
table_id = 'Covid_19_global'
```

```python
to_gbq(df, f"{dataset_id}.{table_id}", project_id=project_id, if_exists='replace')
```
